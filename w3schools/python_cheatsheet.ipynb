```python
# Insert the missing part to loop through the keys and values of all nested dictionaries:
# a = {'name' : 'John', 'age' : 20}
# b = {'name' : 'May', 'age' : 23}
# customers = {'c1' : a, 'c2' : b}
# for x, obj in customers.________:
#   print(x)
    
#   for y in obj:
#     print(y + ':', obj[y])

# answer:
a = {'name' : 'John', 'age' : 20}
b = {'name' : 'May', 'age' : 23}
customers = {'c1' : a, 'c2' : b}
for x, obj in customers.items():
  print(x)
    
  for y in obj:
    print(y + ':', obj[y])
```

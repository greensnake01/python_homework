```python
# input
pressures = [0.273, 0.275, 0.277, 0.275, 0.276] # list

# print
print('pressures:', pressures) # list
print('length:', len(pressures)) # number of values
print()

# fetch an item form a list
print('zeroth item of pressures:', pressures[0])
print('fourth item of pressures:', pressures[4])
print()

# replace an item in a list
pressures[0] = 0.265
print('new pressures:', pressures)
print()

# append items to a list
primes = [2, 3, 5]
print('primes var is initially:', primes)

primes.append(7)
print('primes var has become:', primes)
print()

# extending a list
teen_primes = [11, 13, 17, 19] # list
middle_aged_primes = [37, 41, 43, 48] # list

print('teen primes var is:', teen_primes)
print('middle aged primes var is:', middle_aged_primes)
print('primes var is currently:', primes)

primes.extend(teen_primes)
print('primes var has become:', primes)

primes.extend(middle_aged_primes)
print('primes var has finally become:', primes)
print()

# remove an item from a list
primes = [2, 3, 5, 7, 9] # list
print('primes before removing last item:', primes)

del primes[4] # remove
print('primes after removing last item:', primes)
print()

# an empty list
shopping_list = []
print(shopping_list)
print()

# miscellaneous list
goals = [1, 'Create lists.', 2, 'Extract itmes from lists.',  3, 'Modify lists.']
print(goals)
print()

# 
```

```python
# If start and stop are both non-negative integers, how long is the list values[start:stop]?

countries = ['Germany', 'Poland', 'Russia', 'Italy', 'Hungary']
print(countries[2:5]) # difference = 3; 3 counties are shown
print(countries[1:3]) # difference = 2; 2 countries are shown
print(countries[0:1]) # difference = 1; 1 country is shown

# answer: stop - start, if the stop is greater than the length, the selection will be returned, but shorther than expected
```

```python
# Given this:
print('string to list:', list('tin')) # changes data type into a list and each letter becomes a substring/creates a list with substrings representing each letter separately
print('list to string:', ''.join(['g', 'o', 'l', 'd'])) #joins the letters together forming a string

# What does list('some string') do?
print('string->list: ', list('Wien')) # expects only 1 argument

# answer: a list is formed, accepting only 1 argument (a continuous string of characters), in which every letter becomes a substring

# What does '-'.join(['x', 'y', 'z']) generate? 
print('string formation: ', '-'.join(['x', 'y', 'z'])) # concatenation of each string element in the list and adds the separator between each element in the list

# answer: "x-y-z"
```

```python
# What does the following program print?
element = 'helium'
print(element[-1])

# answer: "m"


# How does Python interpret a negative index?
# answer: negative index position of a string results it backwards reading of the string, starting with "-1" as the first "last" character of the string

# If a list or string has N elements, what is the most negative index that can safely be used with it, and what location does that index represent?
# answer: it is the negative integer defining the length of the a string or a list
print(len(element))
print(element[-6])

# If values is a list, what does del values[-1] do?
# answer: based on the index value and whether it is positive or negative, it deletes a value within the list, e.g. -1 deletes the value from the last position (or first, because '-' turns the 'reading' the other way around, starting from the end)
del values[-1]

# How can you display all elements but the last one without changing values? (Hint: you will need to combine slicing and negative indexing.)
values = [1, 3, 5]
print(values)

# answer: we want to start with everything before the last one and use the last one (-1) on the stop position
print(values[:-1]) 
```

```python
# What does the following program print?
element = 'fluorine'
print(element[::2])
print(element[::-1])

# answer: the first statement prints every second character and the second statement prints everycharacter, but starts from the end, so print the whole string backwards

# If we write a slice as low:high:stride, what does stride do?
print(element[0:5:2])

# answer: [start:stop:step], e.g., [0:5:2] - from zeroth to the one before fifth by every second character

# What expression would select all of the even-numbered items from a collection?
print(element[1::2])

# answer: [1::2]
```

```python
# What do these two programs print? In simple terms, explain the difference between sorted(letters) and letters.sort().

# Program A
letters = list('gold')
result = sorted(letters)
print('letters is', letters, 'and result is', result)

# Program B
letters = list('gold')
result = letters.sort()
print('letters is', letters, 'and result is', result)

# answer: the "Program A" makes a list of characters from the string and sorts them alphabetically, but does not change the initial list (only a copy), while "Program B", sorts the characters alphabetically, although changing the list in the variable "letters" and result prints special data type "None", since it is peculiar to assign an action with a method into a variable
```

```python
# What do these two programs print? In simple terms, explain the difference between new = old and new = old[:].

# Program A
old = list('gold')
new = old      # simple assignment
new[0] = 'D'
print('new is', new, 'and old is', old)

# Program B
old = list('gold')
new = old[:]   # assigning a slice
new[0] = 'D'
print('new is', new, 'and old is', old)

# answer: the "Program A" allows variables "new" and "old" to be assigned to one anotherm thus any change to "new" would also lead to a change in "old", however "Program B" only allows the variable "new" to have inserted a string from the variable "old" 
```

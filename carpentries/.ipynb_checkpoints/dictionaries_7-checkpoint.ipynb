# Dictionaries

## Theory

```python
# dictionary - key: value, (pair)
student_numbers = { 'Bioscience Technology': 16, 
                    'Computational Biology': 12,
                    'Post-Genomic Biology': 20,
                    'Ecology and Environmental Management': 3,
                    'Maths in the Living Environment': 0
                  }

# len()
print('#students:', student_numbers)
print('length:', len(student_numbers))
print()

# fetch via dic[key]
print('#students in Bioscience Technology:', student_numbers['Bioscience Technology'])
print('#students in Post-Genomic Biology:', student_numbers['Post-Genomic Biology'])
print() 

# fetch via .get()
print('#students in Bioscience Technology:', student_numbers.get('Bioscience Technology')) # alternatively
print()

# print keys
print(student_numbers.keys())
print()

# print values
print(student_numbers.values())
print()

# print key: value pair
print(student_numbers.items())
```

```python
# replace/insert a new key: value pair via dic[key] = value
student_numbers['Bioscience Technology'] = 2
print('student_numbers is now:', student_numbers)
print()

# replace/insert a new key: value pari via .update()
student_numbers.update({'Maths in the Living Environment': 120})
print('student_numbers is now:', student_numbers)
```

```python
# dictionary
temperatures = {
    'Monday': 22,
    'Tuesday': 24,
    'Wednesday': 21,
    'Thursday': 23,
    'Friday': 25
}


print('temperatures is initially:', temperatures)
print()

# replace/insert a new key: value pair via dic[key] = value
temperatures['Saturday'] = 24
print('temperatures has become:', temperatures)
print()

# replace/insert a new key: value pari via .update()
temperatures.update({'Sunday': 26})
print('temperatures has become: ', temperatures)
```

```python
temperatures = {
    'Monday': 22,
    'Tuesday': 24,
    'Wednesday': 21,
    'Thursday': 23,
    'Friday': 25,
    'Everyday': 49    
}
print('temperature before removing Everyday:', temperatures)

# delete last entry
del temperatures['Everyday']
print('temperature after removing Everyday:', temperatures)
```

## Miscellaneous

```python
# list
some_list = [4, 3, 8, 9]

# dictionary
some_dict = {
    0: 4,
    1: 3,
    2: 8,
    3: [1, 2, 3]
}

print(some_dict[3]) # access
print(some_dict[3][1]) # specific access
```

```python
# dictionaries keep order of keys as they were added, but in certain versions of python it was not the case, thus this could change and we should not rely on it
shopping_list = {
    # first part is called a 'key', the second is called a 'value'
    'olives': 2, # string or integer: anything,
    'bread': 1,
    'apple': 3
}

shopping_list['olives']
list(shopping_list.keys())[0] # accessing the first key
```

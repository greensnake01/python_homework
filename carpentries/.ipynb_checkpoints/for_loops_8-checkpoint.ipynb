# For Loops

## Theory

```python
# general structure:
# for <loop variable> in <collection>:
#     <body> # indented, starting with the same first letter position as the variable above
```

```python
# simple example
for number in [2, 3, 5]:
    print(number)
```

```python
# more complex example
primes = [2, 3, 5]
for p in primes:
    squared = p ** 2
    cubed = p ** 3
    print(p, squared, cubed)
```

```python
print('a range is not a list: range(0, 3)')
for number in range(0, 3):
    print(number)
```

```python
# Sum the first 10 integers.
total = 0
for number in range(10):
   total = total + (number + 1)
print(total)
```

## Miscellaneous

```python
for num in [1, 2, 3, 4, 5]:
    print(num)
```

```python
s = "the quick brown fox jumps over the lazy dog"

for character in s:
    print(character)

print()

for string in s.split():
    print(string)
```

```python
alphabet = ['a', 'b', 'c']
numbers = [1, 2, 3]
words = ['hello', 'day', 'sun']

for i, letter in enumerate(alphabet):
    print(i, letter, numbers[i], words[i])
```

```python
# Create an acronym: Starting from the list ["red", "green", "blue"], create the acronym "RGB" using a for loop.
# Hint: You may need to use a string method to properly format the acronym.
colours = ['red', 'green', 'blue']
acronym = ''

for colour in colours:
    acronym = acronym + colour[0].upper()
    
print(acronym)
```

```python
# Reorder and properly indent the lines of code below so that they print a list with the cumulative sum of data. The result should be [1, 3, 5, 10].

# original:
# cumulative.append(total)
# for number in data:
# cumulative = []
# total = total + number
# total = 0
# print(cumulative)
# data = [1,2,2,5]

data = [1,2,2,5]
total = 0
cumulative = []

for number in data:
    total = total + number
    cumulative.append(total)

print(cumulative)
```

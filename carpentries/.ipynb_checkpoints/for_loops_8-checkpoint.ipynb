# For Loops

## Theory

```python
# general structure:
# for <loop variable> in <collection>:
#     <body> # indented, starting with the same first letter position as the variable above
```

```python
# use len()
student_numbers = { 'Bioscience Technology': 16, 
                    'Computational Biology': 12,
                    'Post-Genomic Biology': 20,
                    'Ecology and Environmental Management': 3,
                    'Maths in the Living Environment': 0
                  }

print('#students:', student_numbers)
print('length:', len(student_numbers))
print()
print('#students in Bioscience Technology:', student_numbers['Bioscience Technology'])
print('#students in Post-Genomic Biology:', student_numbers['Post-Genomic Biology'])
print() 
print('#students in Bioscience Technology:', student_numbers.get('Bioscience Technology')) # alternatively
```

## Miscellaneous

```python
for num in [1, 2, 3, 4, 5]:
    print(num)
```

```python
s = "the quick brown fox jumps over the lazy dog"

for character in s:
    print(character)

print()

for string in s.split():
    print(string)
```

```python
alphabet = ['a', 'b', 'c']
numbers = [1, 2, 3]
words = ['hello', 'day', 'sun']

for i, letter in enumerate(alphabet):
    print(i, letter, numbers[i], words[i])
```

## Theory

```python
# general structure
# if <condition>:
#     <do something>
# else:
#     <do something>
```

```python
mass = 3.54
if mass > 3.0:
    print(mass, 'is large')

mass = 2.07
if mass > 3.0:
    print (mass, 'is large') # not meeting the condition, thus nothing is printed
```

```python
masses = [3.54, 2.07, 9.22, 1.86, 1.71]
for m in masses:
    if m > 3.0:
        print(m, 'is large')
```

```python
masses = [3.54, 2.07, 9.22, 1.86, 1.71]
for m in masses:
    if m > 3.0:
        print(m, 'is large')
    else:
        print(m, 'is small')
```

```python
masses = [3.54, 2.07, 9.22, 1.86, 1.71]
for m in masses:
    if m > 9.0:
        print(m, 'is HUGE')
    elif m > 3.0:
        print(m, 'is large')
    else:
        print(m, 'is small')
```

```python
velocity = 10.0
if velocity > 20.0:
    print('moving too fast')
else:
    print('adjusting velocity')
    velocity = 50.0
```

```python
velocity = 10.0
for i in range(5): # execute the loop 5 times
    print(i, ':', velocity)
    if velocity > 20.0:
        print('moving too fast')
        velocity = velocity - 5.0
    else:
        print('moving too slow')
        velocity = velocity + 10.0
print('final velocity:', velocity)
```

```python
# Fill in the blanks so that this program creates a new list containing 0s where the original list’s values were negative (-) and 1s where the original list’s values were positive (+).
# original = [-1.5, 0.2, 0.4, 0.0, -1.3, 0.4]
# result = ____
# for value in original:
#     if ____:
#         result.append(0)
#     else:
#         ____

# answer:

##VARIABLES
original = [-1.5, 0.2, 0.4, 0.0, -1.3, 0.4]
result = []

##PROGRAMME
for value in original:
    if value < 0:
        result.append(0)
    else:
        result.append(1)
print(result)
```

```python
# Modify this program so that it finds the largest and smallest values in the list no matter what the range of values originally is.
# values = [...some test data...]
# smallest, largest = None, None
# for v in values:
#     if ____:
#         smallest, largest = v, v
#     ____:
#         smallest = min(____, v)
#         largest = max(____, v)
# print(smallest, largest)

##VARIABLES
values = [1, -10, 375, -769, 1520]
smallest, largest = None, None

##PROGRAMME
# for each value v in values
for v in values:  
    # if accumulator variables are empty
    if smallest == None and largest == None: # apparently 'in' can be used
        # insert v to smallest and largest
        smallest, largest = v, v
    # otherwise
    else:
        # insert smaller compared value into smallest
        smallest = min(smallest, v)
        # insert larger compared value into largest
        largest = max(largest, v)

print('smallest:', smallest, '\nlargest:', largest)

##NOTES
#min/max - if 2 or more arguments are give, returns the smallest/largest 

# answer: 
# advantages: we can see how min/max functions work (most likely); possible to see the process; ???we can define the range within which we would like to work in certain cases (e.g. temperature as a proxy for depth in open oceans of low latitudes - thermocline)??? - not sure; 
# disadvantages: complex; min/max functions exist; time and computational expensive with increasing number of values due to looping through each and min/max functions called within each iteration
```

```python
# without min/max functions
##VARIABLES
values = [-2,1,65,78,-54,-24,100]
smallest, largest = None, None

##PROGRAMME
for v in values:
    # if smallest is none or v is smaller than smallest
    if smallest is None or v < smallest: 
        # insert v into smallest
        smallest = v
    # if largest is none or v is larger than largest
    if largest is None or v > largest:
        # insert v into laargest
        largest = v
print('smallest:', smallest, '\nlargest:', largest)
```

```python
# Trace the values of all variables in this program as it is executed. (Use ‘—’ as the value of variables before and after they exist.)

# global
limit = 100

def clip(value):
    return min(max(0.0, value), limit)

value = -22.5
print(clip(value))

# answer:
# step 1 | limit = 100
# step 2 | limit = 100; definition of a "clip" function - "value" has no value
# step 3 | limit = 100; value = -22.5
# step 4 | limit = 100; value = -22.5; printing result of "clip" function
```

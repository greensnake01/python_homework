### Functions

```python
def get_range(s):
    
    s1 = ''
    s2 = ''
    dash = ''

    for character in range(len(s)):
        if dash == '':
            #print('con1')
            
            if s[character] == '-':
                #print('con1a')
                dash += '-'

            else:
                #print('con1b')
                s1 += s[character]
        
        else:
            #print('con2')
            s2 += s[character]

        if character == len(s)-1:
            #print('clean')
            dash = ''

    return int(s1), int(s2)+1

get_range('11-22') # a, b+1
```

```python
def invalid_mirror(integer):
    
    pattern1 = ''
    pattern2 = ''

    if len(str(integer)) %2 == 0:
        half = len(str(integer)) // 2
        
        for position in range(half): # 0, 1, 2
            pattern1 += str(integer)[position]

        for position in range(half,len(str(integer))): # 3, 4, 5
            pattern2 += str(integer)[position]

        if pattern1 == pattern2:

            return integer

        else: 
            return False
    else:
        return False

invalid_mirror(432432)
```

### Programme

```python
invalid_ID = 0

for item in database:

    for value in range(get_range(item)[0], get_range(item)[1]):

        if value == invalid_mirror(value):
            invalid_ID += value
        else:
            continue
```

## Evaluation of Results

```python
print('sum of invalid IDs:', invalid_ID)
```

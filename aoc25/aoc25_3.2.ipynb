The escalator doesn't move. The Elf explains that it probably needs more joltage to overcome the static friction of the system and hits the big red "joltage limit safety override" button. You lose count of the number of times she needs to confirm "yes, I'm sure" and decorate the lobby a bit while you wait.

Now, you need to make the largest joltage by turning on exactly twelve batteries within each bank.

The joltage output for the bank is still the number formed by the digits of the batteries you've turned on; the only difference is that now there will be 12 digits in each bank's joltage output instead of two.

Consider again the example from before:

987654321111111
811111111111119
234234234234278
818181911112111
Now, the joltages are much larger:

In 987654321111111, the largest joltage can be found by turning on everything except some 1s at the end to produce 987654321111.
In the digit sequence 811111111111119, the largest joltage can be found by turning on everything except some 1s, producing 811111111119.
In 234234234234278, the largest joltage can be found by turning on everything except a 2 battery, a 3 battery, and another 2 battery near the start to produce 434234234278.
In 818181911112111, the joltage 888911112111 is produced by turning on everything except some 1s near the front.
The total output joltage is now much larger: 987654321111 + 811111111119 + 434234234278 + 888911112111 = 3121910778619.

What is the new total output joltage?

### Input

```python
path = '/mnt/c/Users/guspa/Desktop/python_homework/data/aoc25/aoc25_3.1.txt'
mode = 'r'

banks = []

with open(path, mode) as file:
    for line in file.readlines():
        line = line.strip()
        banks.append(line)

print(banks)
```

### Functions

```python
def max_pos(l):

    maximum = max(l)
    maximum_position = l.index(maximum)

    return maximum, int(maximum_position)

max_pos('987654321111111')
```

### Programme

```python
joltage = 0

for battery_series in banks:

    l_battery_series = list(battery_series)

    batteries = 12
    start = 0
    pattern = []

    while batteries > 0:
        end = len(l_battery_series) - (batteries - 1)

        check = l_battery_series[start:end] 

        maximum, maximum_position = max_pos(check)
        maximum_position += start
        start = maximum_position + 1

        pattern.append(maximum)

        batteries -= 1

    joltage += int(''.join(pattern))
```

### Evaluation of Results

```python
print('total necessary joltage:', joltage)
```

### Deconstruction of the Examples

```python
# list with strings as examples
banks = ['987654321111111', '811111111111119', '234234234234278', '818181911112111']
# accumulator variable for sum of "joltages"
joltage = 0

# loop going through all strings in the list
for battery_series in banks:

    l_battery_series = list(battery_series) # transfromation of a string into a list

    batteries = 12 # number of necessary batteries
    start = 0 # starting position
    pattern = [] # accumulator variable for building a pattern

    # while loop going until a 12 digit number is found
    while batteries > 0:
  
        end = len(l_battery_series) - (batteries - 1) # end position until which search takes place; -12 at the beginning to allow for 12 positions to be filled; -1 to allow for movement towards right

        check = l_battery_series[start:end] # search window with size adjusted according to start and stop, but only start position narrows the window; both start and end are adjusted to progress towards right by +1 (start, from starting position) -1 (end, from the last position)

        maximum, maximum_position = max_pos(check) # function returning maximum and its position/index
        maximum_position += start # add starting position
        start = maximum_position + 1 # prepare start position for next iteration by becoming maximum_position (with start position) + 1 (to move by 1 towards right)

        pattern.append(maximum) # maximum is appended into the pattern variable

        batteries -= 1 # makes sure that the window moves by one towards the end

    # construct joltage integer from the pattern
    joltage += int(''.join(pattern))

print('total necessary joltage:', joltage)

# check first 4 at the beginning, starting from the left, because if 3 before were wrong at the beginning then the 4th until the 15th from the 15 digit number would be the number we need; the check is adjusted, if numbers are skipped at the beginning, the search is narrowed accordingly - otherwise 4 positions remain being checked with position moved by 1 to the right
```

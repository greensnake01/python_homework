```python
def invalid_repetition(integer): # e.g. 123123123
    # setup variables
    s = str(integer)
    s_length = len(s) # 9
    code_check = ''

    # for loop going through every digit
    for digit in range(s_length): # e.g 0-8

        # buil a sequence that might repeat within an integer
        sequence = int(s[0:(1+digit)]) #e.g 1; 12; 123
        sequence_length = len(str(sequence)) #e.g 1; 2; 3

        # test if a sequence length could divide the length of an integer 
        if s_length % sequence_length != 0: # 0; continue; 0
            continue
        
        # get number of repetitions
        repetitions = s_length // sequence_length #e.g 9, 3

        # skip an integer which has repetition less than 2 (at least 1 repetition)
        if repetitions < 2: #e.g. 9, 3
            continue
        
        # compares integer with a code which is derived as a result of its division by a found sequence
        if integer % sequence == 0: 
            # get result of integer division by a sequence
            code = integer // sequence #e.g 123123123; 1001001
            pattern = '1' + '0' * (sequence_length-1) #e.g. 1; 100
            code_check = pattern * (repetitions-1) + '1' #e.g. 111111111; 1001001

            # return integer once code and built code match
            if str(code) == code_check:
                return integer

        else:
            continue

    return False

invalid_repetition(123123123)
```

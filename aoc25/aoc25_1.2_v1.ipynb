You're sure that's the right password, but the door won't open. You knock, but nobody answers. You build a snowman while you think.

As you're rolling the snowballs for your snowman, you find another security document that must have fallen into the snow:

"Due to newer security protocols, please use password method 0x434C49434B until further notice."

You remember from the training seminar that "method 0x434C49434B" means you're actually supposed to count the number of times any click causes the dial to point at 0, regardless of whether it happens during a rotation or at the end of one.

Following the same rotations as in the above example, the dial points at zero a few extra times during its rotations:

The dial starts by pointing at 50.
The dial is rotated L68 to point at 82; during this rotation, it points at 0 once.
The dial is rotated L30 to point at 52.
The dial is rotated R48 to point at 0.
The dial is rotated L5 to point at 95.
The dial is rotated R60 to point at 55; during this rotation, it points at 0 once.
The dial is rotated L55 to point at 0.
The dial is rotated L1 to point at 99.
The dial is rotated L99 to point at 0.
The dial is rotated R14 to point at 14.
The dial is rotated L82 to point at 32; during this rotation, it points at 0 once.
In this example, the dial points at 0 three times at the end of a rotation, plus three more times during a rotation. So, in this example, the new password would be 6.

Be careful: if the dial were pointing at 50, a single rotation like R1000 would cause the dial to point at 0 ten times before returning back to 50!

Using password method 0x434C49434B, what is the password to open the door?

### Input

```python
path = '/mnt/c/Users/guspa/Desktop/python_homework/data/aoc25/aoc25_1.1.txt'
mode = 'r'

move_side = []
move_num = []

with open(path, mode) as file:
    for line in file.readlines():
        line = line.strip()

        move_side.append(line[0])

        move_num.append(line[1:])

print(move_side)
print(move_num)
```

### Programme

```python
password = 0
position = 50

for turn in range(len(move_num)):

    # IS STARTING AT 0?
    if position == 0:

        if int(move_num[turn]) >= 100:

            passing_zero = int(move_num[turn]) // 100
 
            password = password + passing_zero
            passing_zero = 0

            move = int(move_num[turn]) % 100
            
        # MOVE IS WITHIN 0-99
        else:
            move = int(move_num[turn])
        
        # IS MOVE L?
        if move_side[turn] == 'L':
            position = position - move

            if position < 0:
                position = position + 100
                
        # MOVE IS R
        else:
            position = position + move

    # STARTING AT 1-99
    else:

        # IS MOVE EXTREME >= 100?
        if int(move_num[turn]) >= 100:

            passing_zero = int(move_num[turn]) // 100

            password = password + passing_zero
            passing_zero = 0

            move = int(move_num[turn]) % 100
        
        # MOVE IS WITHIN 0-99
        else:
            move = int(move_num[turn])
        
        # IS MOVE L?
        if move_side[turn] == 'L':
            position = position - move

            # IS NEGATIVE?
            if position < 0:
                position = position + 100
                password += 1

        # MOVE IS R
        else:
            position = position + move

            # IS LARGER THAN 100?
            if position > 100:
                position = position - 100
                password += 1

            # IS EQUAL TO 100?
            elif position == 100:
                position = position - 100

    # IS FINAL POSITION 0?
    if position == 0:
        password += 1
```

### Evaluation of Results

```python
print('Password:', password)
```

### Deconstruction of the Example

```python
# we need:
# move  | position
# 0     | 50
# L68   | 82 - passed 0
# L30   | 52
# R48   | 0 - is 0
# L5    | 95
# R60   | 55 - passed 0
# L55   | 0 - is 0
# L1    | 99
# L99   | 0 - is 0
# R14   | 14 
# L82   | 32 - passed 0

# password : 6

# what we have:
#move_num = ['68', '30', '48', '5', '60', '55', '1', '99', '14', '82']
#move_side = ['L', 'L', 'R', 'L', 'R', 'L', 'L', 'L', 'R', 'L']

# special cases:
#move_num = ['50', '200', '211', '99']
#move_side = ['L', 'L', 'L', 'L']

# move  | position | password
#       | 50
# L50   | 0        | 1 - end
# L200  | 0        | 3 - passed 0 1x + end
# L211  | 89       | 5 - passed 0 2x
# L99   | 90       | 6- passed 0 1

# accumulator variable for the sums of 0 passed or reached
password = 0
# starting position
position = 50

# for each turn in range of all moves
for turn in range(len(move_num)):

    # IS STARTING AT 0?
    if position == 0:
        # IS MOVE EXTREME >= 100?
        if int(move_num[turn]) >= 100:

            passing_zero = int(move_num[turn]) // 100 # count number of whole circles

            password = password + passing_zero # add counts to password
            passing_zero = 0 # reset counter

            move = int(move_num[turn]) % 100 # use the raminder from division by 100 in another conditions
            
        # MOVE IS WITHIN 0-99
        else:
            move = int(move_num[turn]) # keep the move and use it in another conditions
        
        # IS MOVE L?
        if move_side[turn] == 'L':
            position = position - move

            if position < 0:
                position = position + 100
                

        # MOVE IS R
        else:
            position = position + move

    # STARTING AT 1-99
    else:
        # IS MOVE EXTREME >= 100?
        if int(move_num[turn]) >= 100:

            passing_zero = int(move_num[turn]) // 100 # count number of whole circles

            password = password + passing_zero # add counts to password
            passing_zero = 0 # reset counter

            move = int(move_num[turn]) % 100 # use the raminder from division by 100 in another conditions
        
        # MOVE IS WITHIN 0-99
        else:
            move = int(move_num[turn]) # keep the move and use it in another conditions
        
        # IS MOVE L?
        if move_side[turn] == 'L':
            position = position - move # get new position by subtraction 

            # IS NEGATIVE?
            if position < 0:
                position = position + 100 # add 100 to the position to close the loop
                password += 1 # add 1 to password

        # MOVE IS R
        else:
            position = position + move # get new position by addition

            # IS LARGER THAN 100?
            if position > 100:
                position = position - 100 # subtract 100 from the position to close the loop
                password += 1 # add 1 to password

            # IS EQUAL TO 100?
            elif position == 100:
                position = position - 100 # subtract 100 from the position to close the loop

    # IS FINAL POSITION 0?
    if position == 0:
        password += 1 # add 1 to password

# PASSWORD
print('Password:', password)
```

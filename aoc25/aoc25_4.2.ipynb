## Functions

```python
def scan(position,  dataframe):

    pos_c = position[0]
    pos_r = position[1]

    surrounding_rolls = 0

    if dataframe[pos_c][pos_r] == '@':

        for c in range(-1, 2): # -1, 0, 1
            for r in range(-1, 2): # -1, 0, 1

                if r == 0 and c == 0: # skip centre
                    continue
                
                check_c = pos_c + c
                check_r = pos_r + r
                
                if 0 <= check_r < len(dataframe) and 0 <= check_c < len(dataframe):
                    if dataframe[check_c][check_r] == '@':
                        surrounding_rolls += 1
        
        if surrounding_rolls < 4:
            return True
        else:
            return False
    
    return False

scan([3, 1], test_diagram_cr) # c=3, r=1
```

## Programme

```python
live_diagram_cr = diagram_cr.copy() # test_diagram_cr or diagmra_cr
accessible_rolls = 0

while True:

    positions_to_change = []

    for column in range(len(live_diagram_cr.columns)):
        for row in range(len(live_diagram_cr)):
            
            if scan([column, row], live_diagram_cr) == True:
                    accessible_rolls += 1
                    positions_to_change.append([column, row])
            else: 
                continue

    if not positions_to_change: # if the list is empty, stop
        break

    for position in range(len(positions_to_change)):
        live_diagram_cr.at[positions_to_change[position][1], positions_to_change[position][0]] = 'x'
```

## Deconstruction of the Example

```python
# get a copy of original, but adjusted with borders, diagram to work with
live_diagram_cr = diagram_cr.copy() # test_diagram_cr or diagmra_cr
# integer based accumulator variable for rolls that are accessible - counter
accessible_rolls = 0

# start the while loop
while True:

       # empty list accumulator variable to prevent accumualtion of coordinates being saved
       positions_to_change = []

       # go through each column of a df
       for column in range(len(live_diagram_cr.columns)):
              # go through each row of a df
              for row in range(len(live_diagram_cr)):

                     pos_c = column # save current column position
                     pos_r = row # save current row position
                     surrounding_rolls = 0 # accumulator variable for count of surrounding rolls

                     # check if there is "@" at the current position
                     if live_diagram_cr[pos_c][pos_r] == '@':

                            # check surroundings in 3x3 with current position in the middle
                            for c in range(-1, 2): # -1, 0, 1
                                   for r in range(-1, 2): # -1, 0, 1
                                   
                                          if r == 0 and c == 0: # skip centre, the current position
                                                 continue
                                   
                                          # add positions to current positions from the loops
                                          check_c = pos_c + c
                                          check_r = pos_r + r

                                          # if the position is within the dataframe itself (not outside)
                                          if 0 <= check_r < len(live_diagram_cr) and 0 <= check_c < len(live_diagram_cr):
                                          
                                                 # if there is "@" at one of the positions, add 1 to counter of surrounding rolls
                                                 if live_diagram_cr[check_c][check_r] == '@':
                                                        surrounding_rolls += 1
                     
                            # consider the roll to be accessible if there is less than 4 rolls in the surroundings of 8 position
                            if surrounding_rolls < 4:
                                   # add 1 to counter of easily accessible rolls
                                   accessible_rolls += 1
                                   # add coordinates whenever an accessible roll, which has less than 4 rolls in its surroudnings, is found 
                                   positions_to_change.append([column, row])

       # if the list with coordinates is empty (no more new coordinates), stop
       if not positions_to_change:
              break
       
       # for each coordinates, change the symbol of easily accessible roll to "x"
       for position in range(len(positions_to_change)):
              live_diagram_cr.at[positions_to_change[position][1], positions_to_change[position][0]] = 'x'

# print the number of all easily accessible rolls
print('number of accessible rolls:', accessible_rolls)
```
